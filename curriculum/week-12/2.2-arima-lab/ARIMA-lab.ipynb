```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib as plt
plt.rcParams['figure.figsize'] = (16.0, 8.0)
```

```python
data = pd.read_csv('/Users/macbook/GA-DSI/curriculum/week-12/2.2-arima-lab/assets/datasets/train.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', inplace=True)

data.head()
```

```python


data['Year'] = data.index.year
data['Month'] = data.index.month
```

```python

store1_data = data[data.Store == 1]
```

```python
store1_data.head()
```

```python
weekly_store_sales = store1_data[['Weekly_Sales']].resample('W', how='sum')

pd.rolling_mean(weekly_store_sales, window=3, center=True,).plot()
```

Sales spike around the holidays. The general rise and fall appears to be similar every year just generally increased each year
